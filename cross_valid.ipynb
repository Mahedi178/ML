```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
from sklearn.model_selection import KFold, StratifiedKFold, LeavePOut, ShuffleSplit, LeaveOneOut
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('heart_cleveland_upload.csv')
```

```python
df.head(5)
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
y = df.iloc[:,-1:]
```

```python
y
```

```python
gnb = GaussianNB()
k1_fold = KFold(n_splits = 11)
scores = cross_val_score(gnb, X, y.values.ravel(), cv = k1_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
lr = LogisticRegression(max_iter = 2000)
k2_fold = KFold(n_splits = 27)
scores = cross_val_score(lr, X, y.values.ravel(), cv = k2_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
sv = SVC(kernel='rbf')
k3_fold = KFold(n_splits = 11)
scores = cross_val_score(sv, X, y.values.ravel(), cv = k3_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
rf = RandomForestClassifier()
k4_fold = KFold(n_splits = 9)
scores = cross_val_score(rf, X, y.values.ravel(), cv = k4_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
rf1 = RandomForestClassifier()
sk_folds = StratifiedKFold(n_splits = 9)
scores = cross_val_score(rf1, X, y.values.ravel(), cv = sk_folds)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
rf2 = RandomForestClassifier()
loo = LeaveOneOut()
scores = cross_val_score(rf2, X, y.values.ravel(), cv = loo)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
rf4 = RandomForestClassifier()
ss = ShuffleSplit(train_size=0.6, test_size=0.3, n_splits = 9)
scores = cross_val_score(rf4, X, y.values.ravel(), cv = ss)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
dc = DecisionTreeClassifier()
k5_fold = KFold(n_splits = 11)
scores = cross_val_score(dc, X, y.values.ravel(), cv = k5_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```

```python
clf1 = LogisticRegression(max_iter = 2000)
clf2 = RandomForestClassifier()
clf3 = GaussianNB()
vc = VotingClassifier(estimators=[('RandomForest', clf1), ('Logistics',clf2),('Gaussian',clf3)], voting='hard')
k6_fold = KFold(n_splits = 33)
scores = cross_val_score(vc, X, y.values.ravel(), cv = k6_fold)
print("Cross validation score, ", scores)
print("Average cv scores: ", scores.mean())
print("Number of cv scores used in averages: ", len(scores))
```
