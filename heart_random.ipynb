```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as pt 
```

```python
pd = pd.read_csv('heart_cleveland_upload.csv')
```

```python
pd.isna().sum()
```

```python
x = pd.iloc[:,:-1]
```

```python
x
```

```python
y = pd.iloc[:,-1:]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
for i in range(200):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=i)
    rfc = RandomForestClassifier()
    rfc.fit(xtrain, ytrain.values.ravel())
    dct = DecisionTreeClassifier()
    dct.fit(xtrain, ytrain.values.ravel())
    print(i, rfc.score(xtest, ytest), dct.score(xtest, ytest))
```
