```python
x = pd.iloc[:,:-1]
```

```python
y = pd.iloc[:,-1:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score 
for i in range(200):
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=i)
    rfc = RandomForestClassifier()
    # rfc.fit(xtrain, ytrain.values.ravel())
    # gc = GaussianNB()
    lr = LogisticRegression(max_iter=2000)
    # lr.fit(xtrain, ytrain.values.ravel())
    vc = VotingClassifier(estimators=[('RandomForest', rfc), ('Logistics',lr)], voting='hard')
    vc.fit(xtrain, ytrain.values.ravel())
    v_predict = vc.predict(xtest)
    print(i, accuracy_score(ytest, v_predict))
```
